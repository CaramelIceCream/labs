```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pylab
from matplotlib import mlab
```

```python
N = 2000    #Общая численность населения
I0 = 169     #Начальное количество инфецированных
R0 = 56      #Начально количество людей с иммунитетом
S0 = N - I0 - R0     #Начальное количество восприимчивых
a = 0.01     ##Коэффициент заболеваемости
b = 0.02     #Коэффициент выздоровления
```

```python
v0 = [S0, I0, R0]    #вектор начальных условий
```

```python
t0 = 0
tmax = 200
dt = 0.01
t = np.arange(t0, tmax, dt)
tborders = [t0, tmax]
```

1) Случай, когда I0 <= I*

```python
#система дифференциальных урвнений
def syst(t, x):
    return [0, -b*x[1], b*x[1]]
```

```python
y = solve_ivp(syst, tborders, v0, t_eval=t)   #решение системы на промежутке t
```

```python
plt.figure(figsize=(8, 8))
plt.plot(y.t, y.y.T)
plt.grid()
plt.xlabel('время', fontsize=18)
plt.ylabel('количество особоей', fontsize=18)
plt.legend(['восприимчивые', 'инфицированные', 'выздоровевшие'], shadow=True, fontsize=17)
plt.title('Изменение числа людей в каждой из трех групп ', fontsize=15)
plt.yticks(np.arange(0, 2000, step=100))
```

1) Случай, когда I0 > I*

```python
#система дифференциальных урвнений
def syst(t, x):
    return [-a*x[0], a*x[0]-b*x[1], b*x[1]]
```

```python
t0 = 0
tmax = 600
dt = 0.01
t = np.arange(t0, tmax, dt)
tborders = [t0, tmax]
```

```python
y = solve_ivp(syst, tborders, v0, t_eval=t)   #решение системы на промежутке t
```

```python
plt.figure(figsize=(8, 8))
plt.plot(y.t, y.y.T)
plt.grid()
plt.xlabel('время', fontsize=18)
plt.ylabel('количество особоей', fontsize=18)
plt.legend(['восприимчивые', 'инфицированные', 'выздоровевшие'], shadow=True, fontsize=17)
plt.title('Изменение числа людей в каждой из трех групп ', fontsize=15)
plt.yticks(np.arange(0, 2000, step=100))
```
