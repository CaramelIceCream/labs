```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pylab
from matplotlib import mlab
```

```python
#Вариант 37
M0_1 = 3.8     #начальные средства 1 фирмы
M0_2 = 2.8     #начальные средства 2 фирмы
p_cr = 28      #критическая цена товара
N = 38         #число потребителей
tau1 = 28        #длительность производственного цикла 1 фирмы
tau2 = 18        #длительность производственного цикла 2 фирмы
p1 = 8.8       #себестоимость товара 1 фирмы
p2 = 11.8      #себестоимость товара 2 фирмы
q = 1          #максимальная потребность одного человека в продукте в единицу времени
```

```python
#Коэффициенты для системы
a1 = p_cr/(tau1*tau1*p1*p1*N*q);
a2 = p_cr/(tau2*tau2*p2*p2*N*q);
b = p_cr/(tau1*tau1*tau2*tau2*p1*p1*p2*p2*N*q);
c1 = (p_cr-p1)/(tau1*p1);
c2 = (p_cr-p2)/(tau2*p2);
```

```python
V0 = [M0_1, M0_2]     #вектор начальных условий
t0 = 0
tmax = 30
dt = 0.01
t = np.arange(t0, tmax, dt)
tborders = [t0, tmax]
```

```python
#система дифференциальных урвнений для 1 случая
def syst(t, M):
    return [M[0] - (b/c1)*M[0]*M[1] - (a1/c1)*M[0]*M[0], (c2/c1)*M[1] - (b/c1)*M[0]*M[1] - (a2/c1)*M[1]*M[1]]
```

```python
M = solve_ivp(syst, tborders, V0, t_eval=t)   #решение системы на промежутке t
```

```python
plt.figure(figsize=(8, 8))
plt.plot(M.t, M.y.T)
plt.grid()
plt.xlabel('Время', fontsize=18)
plt.ylabel('Оборотные средства', fontsize=18)
plt.legend(['фирма 1', 'фирма 2'], shadow=True, fontsize=17)
plt.title('Изменение оборотных средств фирмы 1 и фирмы 2', fontsize=15)
plt.yticks(np.arange(0, 7000, step=500))
plt.xticks(np.arange(0, 30, step=2))
```

```python
#система дифференциальных урвнений для 2 случая
def syst(t, M):
    return [M[0] - (b/c1)*M[0]*M[1] - (a1/c1)*M[0]*M[0], (c2/c1)*M[1] - (b/c1 + 0.00063)*M[0]*M[1] - (a2/c1)*M[1]*M[1]]
```

```python
M = solve_ivp(syst, tborders, V0, t_eval=t)   #решение системы на промежутке t
```

```python
plt.figure(figsize=(8, 8))
plt.plot(M.t, M.y.T)
plt.grid()
plt.xlabel('Время', fontsize=18)
plt.ylabel('Оборотные средства', fontsize=18)
plt.legend(['фирма 1', 'фирма 2'], shadow=True, fontsize=17)
plt.title('Изменение оборотных средств фирмы 1 и фирмы 2', fontsize=15)
plt.yticks(np.arange(0, 7000, step=500))
plt.xticks(np.arange(0, 30, step=2))
```
