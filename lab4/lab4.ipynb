```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math
```

```python
t=np.arange(0, 68, 0.05)
x0=1.8
y0=0.8
v0=[x0, y0]     #вектор начальных условий
```

```python
w2 = 6
tmax = 68
step = 0.05
y0 = [0, 0]
```

<b>1 случай. Колебания гармонического осциллятора без затуханий и без действий внешней
силы</b>

```python
def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 ]
```

```python
t = np.arange( 0, tmax, step)
w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]
```

```python
fig = plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
```

```python
fig2 = plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
```

<b>2 случай. Колебания гармонического осциллятора c затуханием и без действий внешней

```python
## 2 
w2 = 6
g = 6
```

```python
def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 - g*y2 ]
```

<b>3 случай. Колебания гармонического осциллятора с затуханием и действием внешней
силы</b>

```python
## 3
w2 = 6
g = 12
```

```python
def f(t): 
    f = math.sin(6*t)
    return f
```

```python
def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 - g*y2 + f(t) ]
```

```python
t = np.arange( 0, tmax, step)
w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]
```

```python
fig = plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
```

```python
fig2 = plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
```
