```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s=19.1 #начальное расстояние лодки от катера
fi=3*math.pi/4
```

```python
def f(tetha, r):
    return r/math.sqrt(15.21)
```

```python
r0=s/2.9
tetha0=-math.pi
```

```python
tetha = np.arange(-math.pi, 2*math.pi, 0.01)
#tetha.insert(0, math.pi)
```

```python
r = odeint(f, r0, tetha)
```

```python
#for i, val in enumerate(tetha):
#    tetha[i] = val*math.pi
```

```python
#функция, описывающая движение лодки браконьеров
def f2(t):
    y = [math.tan(fi)*i for i in t]
    return y
```

```python
t = np.arange(0, 1000, 1)
```

```python
plt.plot(t,  f2(t))
```

```python
plt.plot(t,  f2(t))
fig = plt.figure(figsize=(8., 6.))
ax1 = fig.add_subplot(231, projection='polar')
ax1.plot(tetha, r)
```

```python
fig = plt.figure()
rect = [0.1, 0.1, 0.8, 0.8]
ax_carthesian  = fig.add_axes(rect)
ax_polar = fig.add_axes(rect, polar=True, frameon=False)
ax_carthesian.plot(t, f2(t),'b')
xax = ax_carthesian.xaxis
xax.set_view_interval(-2000, 2000)
yax = ax_carthesian.yaxis
yax.set_view_interval(-2000, 2000)

ax_polar.plot(tetha, r, color='r', linewidth=3)
#ax_polar.set_rmax(2.0)
ax_polar.grid(True)
```
