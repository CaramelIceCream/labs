### Вариант 37

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
N = 1140     #объем аудитории
N0 = 10     #количество людей, знающих о товаре в начальный момент времени
```

## Первый случай

```python
alpha1 = 0.13     #интенсивность рекламной кампании
alpha2 = 0.000013  #сарафанное радио

t = np.arange(0, 8, 0.1)     #временной промежуток(длительность рекламной кампании)

#уравнение, описывающее распространение рекламы
def xd(t, x):
    return ( alpha1 + alpha2*x )*( N - x )

x = odeint(xd, N0, t)

plt.plot(t,  x)
plt.xlabel("Время")
plt.ylabel("численность")
plt.grid(True)
#plt.yticks(np.arange(0, N, step=50))
```

## Второй случай

```python
alpha1 = 0.000031     #интенсивность рекламной кампании
alpha2 = 0.31  #сарафанное радио

t = np.arange(0, 3, 0.1)     #временной промежуток(длительность рекламной кампании)

#уравнение, описывающее распространение рекламы
def xd(t, x):
    return ( alpha1 + alpha2*x )*( N - x )

x = odeint(xd, N0, t)

dx = (alpha1 + x*alpha2)*(N-x)

plt.plot(t,  x)
#plt.plot(t, dx)
plt.xlabel("время")
plt.ylabel("численность")
plt.grid(True)
```

## Третий случай

```python
#N = 200

#интенсивность рекламной кампании
def alpha1(t):
    return 0.13*t

#сарафанное радио
def alpha2(t):
    return 0.31*np.cos(t)

t = np.arange(0, 3, 0.01)     #временной промежуток(длительность рекламной кампании)

#уравнение, описывающее распространение рекламы
def xd(t, x):
    return ( alpha1(t) + alpha2(t)*x )*( N - x )

x = odeint(xd, N0, t)


plt.figure(figsize=(8, 8))
plt.plot(t,  x)
plt.xlabel("время")
plt.ylabel("численность")
plt.grid(True)
#plt.yticks(np.arange(0, N, step=10))
```
